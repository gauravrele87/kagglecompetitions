```python
%matplotlib inline

import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
random.seed(1)
x = np.random.rand(50)
c = 2.3
m = 1.5

y = m*x + c
```

```python
def predicted_function(x,w0,w1):
    return w0+x*w1
```

```python
from matplotlib import *
from matplotlib.animation import FuncAnimation
%matplotlib notebook
plt.rcParams["animation.html"] = "jshtml"


fig = plt.figure()
ax = plt.axes(xlim=(0, 1), ylim=(0, 5))
plt.plot(x,y,'.')
line, = ax.plot([], [])
plt.close()

guess_w0 = -1
guess_w1 = -1

def init():
    line.set_data(x,y)
    #line.set_marker('.')
    #line.set_linestyle(' ')

def gradient_descent(i,guess_w0,guess_w1,x):
    lr = 0.1
    for _ in range(i):
        dLdw0 = -2*np.sum(y - predicted_function(x,guess_w0,guess_w1))/float(len(x))
        dLdw1 = -2*np.sum((y - predicted_function(x,guess_w0,guess_w1))*x)/float(len(x))
        guess_w0 -= lr* dLdw0
        guess_w1 -= lr* dLdw1
        line.set_data(x, guess_w0 + x*guess_w1)
    return line,

    
#anim = FuncAnimation(fig, animate, init_func=init, frames=200,fargs=(guess_w0,guess_w1,x),interval=10)
anim = FuncAnimation(fig, gradient_descent, init_func=init,fargs=(guess_w0,guess_w1,x), frames=200,interval=20)
anim
from IPython.display import HTML
HTML(anim.to_jshtml())
```
